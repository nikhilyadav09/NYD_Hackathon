```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage
url = "https://patanjaliyogasutra.in/samadhipada1-2/"

# Define headers to mimic a browser
headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
        "(KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36"
    ),
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Connection": "keep-alive",
}

# Fetch the page with headers
response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")

    # Find all elements with the class `tabs-cont-box active`
    tabs_content = soup.find_all(class_="tabs-cont-box active")

    # Extract and print the content
    for i, tab in enumerate(tabs_content, 1):
        print(f"Tab {i}:")
        print(tab.get_text(strip=True))  # Extract text and strip extra whitespace
        print("-" * 50)
else:
    print(f"Failed to fetch the page. Status code: {response.status_code}")
```

```python
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import time

# Initialize the browser
options = uc.ChromeOptions()
options.add_argument("--headless")  # Run in headless mode
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Output CSV file
output_file = "patanjali_translations.csv"

# Initialize CSV file with headers
with open(output_file, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Chapter", "Verse", "Translation"])  # Headers

# Function to extract content from a specific URL
def extract_translation(driver, url, chapter, verse):
    try:
        driver.get(url)

        # Wait for the tab buttons to load
        wait = WebDriverWait(driver, 10)
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, "tabsbtn-wrapper")))

        # Click on the "English" tab
        english_tab = driver.find_element(By.XPATH, "//li[text()='English']")
        english_tab.click()

        # Wait for the English content to load
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, "tabs-cont-box.active")))

        # Extract the English translation
        content = driver.find_element(By.CLASS_NAME, "tabs-cont-box.active").text

        print(f"Chapter {chapter}, Verse {verse} extracted successfully.")
        return content

    except Exception as e:
        print(f"Failed to extract data for Chapter {chapter}, Verse {verse}. Error: {e}")
        return None

# Main script
def scrape_patanjali_translations():
    driver = uc.Chrome(options=options)

    try:
        chapter = 1
        while chapter <= 18:  # Loop through chapters
            verse = 1
            while True:  # Loop through verses in a chapter
                url = f"https://patanjaliyogasutra.in/samadhipada{chapter}-{verse}/"
                print(f"Processing URL: {url}")

                # Extract translation
                translation = extract_translation(driver, url, chapter, verse)
                if translation is None or translation.strip() == "":
                    print(f"No data found for Chapter {chapter}, Verse {verse}. Moving to the next chapter.")
                    break  # Move to the next chapter

                # Append data to the CSV file
                with open(output_file, mode="a", newline="", encoding="utf-8") as file:
                    writer = csv.writer(file)
                    writer.writerow([chapter, verse, translation])

                # Increment verse
                verse += 1

            # Increment chapter
            chapter += 1

    finally:
        driver.quit()
        print("Scraping completed.")

# Run the script
if __name__ == "__main__":
    scrape_patanjali_translations()
```
